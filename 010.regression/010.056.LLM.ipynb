````python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import Lasso
from sklearn.metrics import r2_score
from sklearn.model_selection import KFold
from sklearn.datasets import load_diabetes


# ハイパーパラメータの設定
alpha_list = np.logspace(-5, 2, 20)

# 交差検証の設定
kf = KFold(n_splits=5, shuffle=True, random_state=42)

# 結果を保存するためのリスト
train_scores = []
test_scores = []
mean_train_scores = []
mean_test_scores = []
std_train_scores = []
std_test_scores = []

# ハイパーパラメータのループ
for alpha in alpha_list:
    temp_train_scores = []
    temp_test_scores = []

    # 交差検証ループ
    for train_index, test_index in kf.split(X):
        X_train, X_test = X[train_index], X[test_index]
        y_train, y_test = y[train_index], y[test_index]

        # モデルのインスタンス化と訓練
        model = Lasso(alpha=alpha)
        model.fit(X_train, y_train)

        # スコアの計算
        y_train_pred = model.predict(X_train)
        y_test_pred = model.predict(X_test)
        temp_train_scores.append(r2_score(y_train, y_train_pred))
        temp_test_scores.append(r2_score(y_test, y_test_pred))

    # スコアの集計
    train_scores.append(temp_train_scores)
    test_scores.append(temp_test_scores)
    mean_train_scores.append(np.mean(temp_train_scores))
    mean_test_scores.append(np.mean(temp_test_scores))
    std_train_scores.append(np.std(temp_train_scores))
    std_test_scores.append(np.std(temp_test_scores))

# 結果の可視化
plt.figure(figsize=(10, 6))
plt.errorbar(alpha_list, mean_train_scores, yerr=std_train_scores, label='Train Score', fmt='-o')
plt.errorbar(alpha_list, mean_test_scores, yerr=std_test_scores, label='Test Score', fmt='-o')
plt.xscale('log')
plt.xlabel('Alpha')
plt.ylabel('R^2 Score')
plt.title('R^2 Scores for different Alpha values in Lasso Regression')
plt.legend()
plt.show()
````

from sklearn.linear_model import Lasso
import matplotlib.pyplot as plt
import numpy as np

# 仮のデータを生成（実際にはユーザーが提供するデータを使用する）
np.random.seed(0)
X = np.random.randn(100, 3)
y = X @ np.array([1.5, -2., 3.]) + np.random.randn(100) * 0.5

# alphaのリスト（ユーザーが提供する値を使用する）
alpha_list = [0.1, 0.5, 1, 5, 10]

# 各alphaに対する回帰係数を格納するリスト
coefs = []

for alpha in alpha_list:
    lasso = Lasso(alpha=alpha)
    lasso.fit(X, y)
    coefs.append(lasso.coef_)

# 可視化
plt.figure(figsize=(10, 6))
ax = plt.gca()

ax.plot(alpha_list, coefs)
ax.set_xscale('log')
plt.xlabel('alpha')
plt.ylabel('coefficients')
plt.title('Lasso coefficients as a function of the regularization')
plt.axis('tight')
plt.legend(['Feature 1', 'Feature 2', 'Feature 3'])

````python
from sklearn.linear_model import Lasso

# 各alphaに対する回帰係数を格納するためのリスト
coefficients = []

# Lasso回帰モデルの学習と回帰係数の収集
for alpha in alpha_list:
    lasso = Lasso(alpha=alpha, max_iter=10000, tol=0.01)
    lasso.fit(X, y)
    coefficients.append(lasso.coef_)

# 回帰係数の可視化
plt.figure(figsize=(15, 6))
for i in range(X.shape[1]):
    plt.plot(alpha_list, [coef[i] for coef in coefficients], label=f'Coefficient {i+1}')

plt.xscale('log')
plt.xlabel('Alpha')
plt.ylabel('Coefficients')
plt.title('Lasso Coefficients as a function of the Regularization')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
````
