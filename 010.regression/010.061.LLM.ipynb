````python
from sklearn.model_selection import GridSearchCV, KFold, cross_val_score
from sklearn.kernel_ridge import KernelRidge
import matplotlib.pyplot as plt
import numpy as np


# モデルの設定
model = KernelRidge(kernel='rbf')

# ハイパーパラメータの最適化
param_grid = {"alpha": [1e0, 1e-1, 1e-2, 1e-3],
              "gamma": np.logspace(-2, 2, 5)}
grid = GridSearchCV(model, param_grid=param_grid)
grid.fit(X, y)
print(f'Best parameters: {grid.best_params_}')

# 最適なハイパーパラメータでモデルを設定
model = KernelRidge(kernel='rbf', **grid.best_params_)

# 交差検証
cv = KFold(n_splits=10, shuffle=True, random_state=0)
scores = cross_val_score(model, X, y, cv=cv, scoring='r2')
print(f'Average R2 score: {np.mean(scores)}')

# モデルのフィッティング
model.fit(X, y)

# 予測
predictions = model.predict(X)

# プロット
plt.scatter(y, predictions)
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title('Actual vs Predicted Values')
plt.show()
````

````python
from sklearn.model_selection import GridSearchCV, KFold, cross_val_score
from sklearn.kernel_ridge import KernelRidge
import matplotlib.pyplot as plt
import numpy as np


# モデルの設定
model = KernelRidge(kernel='polynomial')

# ハイパーパラメータの最適化
param_grid = {"alpha": [1e0, 1e-1, 1e-2, 1e-3],
              "gamma": np.logspace(-2, 2, 5),
              "degree": [2, 3, 4, 5]}
grid = GridSearchCV(model, param_grid=param_grid)
grid.fit(X, y)
print(f'Best parameters: {grid.best_params_}')

# 最適なハイパーパラメータでモデルを設定
model = KernelRidge(kernel='polynomial', **grid.best_params_)

# 交差検証
cv = KFold(n_splits=10, shuffle=True, random_state=0)
scores = cross_val_score(model, X, y, cv=cv, scoring='r2')
print(f'Average R2 score: {np.mean(scores)}')

# モデルのフィッティング
model.fit(X, y)

# 予測
predictions = model.predict(X)

# プロット
plt.scatter(y, predictions)
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title('Actual vs Predicted Values')
plt.show()
````

````python
from sklearn.model_selection import GridSearchCV, KFold, cross_val_score
from sklearn.kernel_ridge import KernelRidge
import matplotlib.pyplot as plt
import numpy as np


# モデルの設定
model = KernelRidge(kernel='polynomial')

# ハイパーパラメータの最適化
param_grid = {"alpha": [1e0, 1e-1, 1e-2, 1e-3],
              "gamma": np.logspace(-2, 2, 5),
              "degree": [2, 3, 4, 5],
             "coef0": [0,1]}
grid = GridSearchCV(model, param_grid=param_grid, verbose=4)
grid.fit(X, y)
print(f'Best parameters: {grid.best_params_}')

# 最適なハイパーパラメータでモデルを設定
model = KernelRidge(kernel='polynomial', **grid.best_params_)

# 交差検証
cv = KFold(n_splits=10, shuffle=True, random_state=0)
scores = cross_val_score(model, X, y, cv=cv, scoring='r2')
print(f'Average R2 score: {np.mean(scores)}')

# モデルのフィッティング
model.fit(X, y)

# 予測
predictions = model.predict(X)

# プロット
plt.scatter(y, predictions)
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title('Actual vs Predicted Values')
plt.show()
````
