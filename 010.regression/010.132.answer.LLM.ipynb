````python
# file_pathを直す。

import pandas as pd
import numpy as np
from sklearn.model_selection import cross_val_score, KFold
from sklearn.linear_model import Lasso
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# Load the data
file_path = '../data_calculated/x5_sin.csv'
df_obs = pd.read_csv(file_path)

# Defining the descriptor and target names
DESCRIPTOR_NAMES = ['x1', 'x2', 'x3', 'x4', 'x5', 'x6']
TARGET_NAME = "y"

# Extracting the features and target
Xraw = df_obs[DESCRIPTOR_NAMES]
y = df_obs[TARGET_NAME]

# Standardization of explanatory variables
scaler = StandardScaler()
X = scaler.fit_transform(Xraw)

# Lasso hyperparameter optimization
lasso_hyperparams = np.logspace(-5, 2, 20)
kf = KFold(n_splits=10, shuffle=True, random_state=42)

best_score = -np.inf
best_alpha = None
for alpha in lasso_hyperparams:
    lasso = Lasso(alpha=alpha, random_state=42)
    scores = cross_val_score(lasso, X, y, cv=kf, scoring='r2')
    avg_score = np.mean(scores)
    if avg_score > best_score:
        best_score = avg_score
        best_alpha = alpha

# Train the model with the best hyperparameter
lasso_best = Lasso(alpha=best_alpha, random_state=42)
lasso_best.fit(X, y)

# Calculate R2 score and predictions
r2_score = lasso_best.score(X, y)
yp = lasso_best.predict(X)

# y vs yp plot
plt.figure(figsize=(8, 6))
plt.scatter(y, yp, edgecolor='k', alpha=0.7)
plt.xlabel('Actual y')
plt.ylabel('Predicted yp')
plt.title('Actual y vs Predicted yp')
plt.grid(True)
plt.show()

best_alpha, r2_score
````
