````python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import numpy as np

# PCAのインスタンスを作成し、2次元に次元圧縮します。
pca = PCA(n_components=2)
Xdr = pca.fit_transform(X)

# 次元圧縮後の変数Xdrを二次元で図示します。
plt.figure(figsize=(6, 6))

# splabelsの名前毎にXdrの色を変え、legendを書きます。
labels = np.unique(splabels)
for label in labels:
    plt.scatter(Xdr[splabels == label, 0], Xdr[splabels == label, 1], label=label)

plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
plt.title('PCA Dimensionality Reduction to 2D')
plt.legend()
plt.show()
````

````python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import numpy as np

# PCAのインスタンスを作成し、2次元に次元圧縮します。
model = PCA(n_components=2)
Xdr = model.fit_transform(X)

# 説明変数Xnewに対してmodelを適用してXdr_newを作成します。
# ここでは、ダミーデータを使用します。
# 実際には、あなたのデータセットに合わせてXnewを定義してください。
Xdr_new = model.transform(Xnew)

# 次元圧縮後の変数Xdr_newを二次元で図示します。
plt.figure(figsize=(6, 6))
plt.scatter(Xdr_new[:, 0], Xdr_new[:, 1], color='gray', alpha=0.3)

# splabelsの名前毎にXdrの色を変え、legendを書きます。
labels = np.unique(splabels)
for label in labels:
    plt.scatter(Xdr[splabels == label, 0], Xdr[splabels == label, 1], label=label)

plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
plt.title('PCA Dimensionality Reduction to 2D')
plt.legend()
plt.show()
````
