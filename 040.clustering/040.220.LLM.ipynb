````python
import numpy as np
import pandas as pd
from scipy.cluster.hierarchy import dendrogram, linkage
import matplotlib.pyplot as plt

def plot_dendrogram(X, descriptor_names):
    # numpy配列からpandas DataFrameを作成
    df = pd.DataFrame(X, columns=descriptor_names)

    # Pearsonの相関行列を計算
    corr_matrix = df.corr()

    # 相関行列から距離行列を計算
    distance_matrix = 1 - corr_matrix

    # 階層的クラスタリング（complete linkage）
    Z = linkage(distance_matrix, 'complete', optimal_ordering=True)

    # デンドログラムの描画
    plt.figure(figsize=(10, 7))
    dendrogram(Z, labels=descriptor_names, orientation='top')
    plt.title("Hierarchical Clustering Dendrogram")
    plt.xlabel("Feature")
    plt.ylabel("Distance")
    plt.show()

# 使用例
# X = np.array(...) # ここでXにデータを提供
# descriptor_names = [...] # ここでdescriptor_namesに名前のリストを提供
plot_dendrogram(X, descriptor_names)
````
