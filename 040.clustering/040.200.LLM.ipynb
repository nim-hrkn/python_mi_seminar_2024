````python
from sklearn.mixture import GaussianMixture
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# 変数XをPCAで二次元に変換
pca = PCA(n_components=2)
Xpca = pca.fit_transform(X)

# GaussianMixtureでXpcaを３つにクラスタリング
gmm = GaussianMixture(n_components=3)
gmm.fit(Xpca)
yp = gmm.predict(Xpca)

# yp毎にXpcaの散布図を書く
plt.figure(figsize=(10, 7))
for i in range(3):
    plt.scatter(Xpca[yp == i, 0], Xpca[yp == i, 1], label=f'Cluster {i+1}')
plt.legend()
plt.show()
````

````python
def transform_point(x, y, cov_matrix):
    # Compute eigenvectors and eigenvalues
    eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)

    # Normalize the eigenvectors
    eigenvectors = eigenvectors / np.linalg.norm(eigenvectors, axis=0)

    # Transform the point
    transformed_point = np.dot(eigenvectors.T, np.array([x, y]))

    return transformed_point

cov_matrix = [[1.23012608, -0.96532353],
              [-0.96532353, 0.91748451]]

````

````python
from scipy.stats import multivariate_normal

def gaussian_distribution_scipy(x, xmean, cov):
    """
    Compute the probability density of the multivariate Gaussian distribution using scipy.

    :param x: Input vector (numpy array).
    :param xmean: Mean vector of the Gaussian distribution.
    :param cov: Covariance matrix of the Gaussian distribution.
    :return: Probability density at the given point x.
    """
    # Create a multivariate normal distribution
    mvn = multivariate_normal(mean=xmean, cov=cov)

    # Compute the PDF
    pdf = mvn.pdf(x)

    return pdf

````

````python
Xtr = transform_point(Xpca[:,0], Xpca[:,1], cov_matrix).T

# yp毎にXpcaの散布図を書く
plt.figure(figsize=(10, 7))
for icenter in range(3):
    # Example usage
    xmean = gmm.means_[icenter]
    cov = gmm.covariances_[icenter]
    pdf_value = gaussian_distribution_scipy(Xtr, xmean, cov) *gmm.weights_[icenter]

    plt.scatter( Xtr[:, 1], pdf_value, label=f'Cluster {icenter+1}', s=10, alpha=0.5)
plt.legend()
plt.show()
````
